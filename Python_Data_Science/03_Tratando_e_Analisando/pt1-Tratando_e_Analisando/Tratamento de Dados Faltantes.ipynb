## Tratamento de Dados Faltantes (continuação)

```
# Número de dados que contém condomínio sem informação
dados[dados['Condominio'].isnull()].shape[0]
```

```
selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
```

```
A = dados.shape[0]
dados = dados[~selecao]
B = dados.shape[0]
A - B
```

```
dados[dados['Condominio'].isnull()].shape[0]
```

```
# Pode incluir mais de uma variável dessa forma. 
# utilizando ['Condominio'], inplace=True precisaria um por vez para cada variável
dados = dados.fillna({'Condominio': 0, 'IPTU': 0})
```
