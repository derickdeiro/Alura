# Estatística Descritivas

```
# mostra uma descrição estatística com quantidade, média, desvio padrão, mínimo, 25%, 50%, 75% e máximo.
grupo_bairro['Valor'].describe().round(2)
```

```
# Aggregate permite criar apenas as informações que você deseja visualizar
grupo_bairro['Valor'].aggregate(['min', 'max']).rename(columns = {'min': 'Mínimo', 'max': 'Máximo'})
```

```
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (20, 10))
```

```
# gráfico para mostar o desvio padrão dos bairros
fig = grupo_bairro['Valor'].std().plot.bar()
fig.set_ylabel('Desvio Padrão do Aluguel')
fig.set_title('Desvio Padrão do Alguel por Bairro', {'fontsize': 22})
```

```
fig = grupo_bairro['Valor'].mean().plot.bar()
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor Médio do Alguel por Bairro', {'fontsize': 22})
```

```
fig = grupo_bairro['Valor'].max().plot.bar()
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor Máximo do Alguel por Bairro', {'fontsize': 22})
```
