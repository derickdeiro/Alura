```
import pandas as pd
```

```
data = [0.5, None, None, 0.52, 0.54, None, None, 0.59, 0.6, None, 0.7]
s = pd.Series(data)
```

```
s
```

```
s.fillna(0)
```

```
s.fillna(method= 'ffill')  # preenche o dados com o dado anterior
```

```
s
```

```
s.fillna(method='bfill')  # preenche com os dados posteriores
```

```
s.fillna(s.mean())  # preenche com a média dos valores
```

```
s
```

```
# Preenche apenas o primeiro dado encontrado com o valor anterior, por conta do Limit=1
s1 = s.fillna(method= 'ffill', limit=1)
s1
```

```
# intercalando com o primeiro valor e com o segundo para preencher
s1 = s1.fillna(method='bfill', limit=1)
s1
```
