## Seleções e Frequências

```
import pandas as pd
```

```
dados = pd.read_csv('/content/drive/MyDrive/Alura/Curso Pandas/aluguel_residencial.csv', sep=';')
```

```
dados
```

```
# Selecione somente os imóveis classificados com tipo "Apartamento"
selecao = dados['Tipo'] == 'Apartamento'
registros_apartamentos = dados[selecao].shape[0]
registros_apartamentos
```

```
# Selecione os imóveis classificados com tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila'
selecao_casa = (dados['Tipo'] == 'Casa') | (dados['Tipo'] == 'Casa de Condomínio') | (dados['Tipo'] == 'Casa de Vila')
registros_casas = dados[selecao_casa].shape[0]
registros_casas
```

```
# imóveis imóveis com área entre 60 e 100 metros quadrados, incluindo os limites
selecao_mt_quadrado = (dados['Area'] >= 60) & (dados['Area'] <= 100)
registros_mt_quadrado = dados[selecao_mt_quadrados].shape[0]
registros_mt_quadrado
```

```
# Selecione os imoveis que tenham pelo o menos 4 quartos e aluguel menor que R$ 2.000,00
selecao_quartos = (dados['Quartos'] >= 4) & (dados['Valor'] < 2000.0)
registros_quartos = dados[selecao_quartos].shape[0]
registros_quartos
```

```
print(f'Nº de imóveis classificados com tipo "Apartamento" -> {registros_apartamentos}')
print(f'Nº de imóveis classificados com tipos "Casa", "Casa de Condomínio" e "Casa de Vila" -> {registros_casas}')
print(f'Nº de imóveis com área entre 60 e 100 metros quadrados, incluindo os limites -> {registros_mt_quadrado}')
print(f'Nº de imoveis que tenham pelo o menos 4 quartos e aluguel menor que R$ 2.000,00 -> {registros_quartos}')
```
